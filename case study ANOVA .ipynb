# Q3.Here all monthly sales data is compared with individual zones , hence one way anova is done  

```python
#Create the variables
A=df['Zone - A']
B=df['Zone - B']
C=df['Zone - C']
D=df['Zone - D']
```

# H_0 : mean_zone A = mean_zone B = mean_ zone C= mean_zone D ; implies all zone have same sales volume.

# H_1 : atleast one mean is different

#Calculating the F-statistics and p-value



```python
F,p=stat.f_oneway(A,B,C,D,axis=0)
print('F-value: ',F)
print('p-value : ',p)
```

# Here p-value < 0.05 .
Hence we reject the null hypothesis.
Therefore there is significant difference between the means of the sales vol. of different zones .
