# K-Means clustering model

```python
from sklearn.cluster import KMeans
dt= range(1,13)
wcss=[]
for d in dt:
   model=KMeans(n_clusters=d,init='k-means++',random_state=42)
   model.fit(data)
   wcss.append(model.inertia_)
```

# plotting the elbow for cluster dependancy number analysis

```python
plt.plot(range(1,13),wcss)
plt.xlabel('the no of cluster')
plt.ylabel('wccs')
plt.show()
```

Here the cluster number clearly is = 4 from the elbow diagram , hence n_cluster =4 is considered.

# K-Means modelling with the cluster number from elbow method

```python
data_model=KMeans(n_clusters=4,init='k-means++',random_state=101)
d_means=data_model.fit(data)
labelkm=data_model.predict(data)
print(labelkm)
```

# checking the value dependency with silhouette score

```python
from sklearn.metrics import silhouette_score
c=silhouette_score(data, labelkm, metric='euclidean')
print(c,':The value is in between 0&1, also the value can be treated as a fair means of considering the clustering')
```

# hierarchial clustering

```python
#importing dendrogram and linkage function from scipy
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
#plotting the dendrogram for the data, specifying with the intersect line
datalink=linkage(data,method='ward')
plt.figure(figsize=(10,10))
dendrogram(datalink,orientation='top',distance_sort='ascending',show_leaf_counts=True)
plt.hlines(2000,0,3000,linestyles='dashed',color='brown')
plt.show()
```

 As the  hline from the largest line intersects 3 lines , the number of clusters values to be considered 3, for agglomerative modelling

```python
#creating the instance and predicting and fitting the data based on agglomertive clustering. 
from sklearn.cluster import AgglomerativeClustering
datacluster=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
labelagglo=datacluster.fit_predict(data)
labelagglo
```

```python
#to check the dependablity of the clustered data , inertia value is re checked with silhoutte score.
s=silhouette_score(data, labelagglo, metric='euclidean')
print(s,':The value is in between 0&1, also the value can be treated as a fair means of considering the clustering')
```
