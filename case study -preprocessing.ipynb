```python
Q1=np.percentile(data['Fare'],25,interpolation='midpoint')
Q2=np.percentile(data['Fare'],50,interpolation='midpoint')
Q3=np.percentile(data['Fare'],75,interpolation='midpoint')
print(Q1)
print(Q2)
print(Q3)
```

```python
IQR=Q3-Q1
low=Q1-1.5*IQR
upper=Q3+1.5*IQR
print(low)
print(upper)
```

```python
outlier=[]
for a in data['Fare']:
    if((a>upper)or(a<low)):
        outlier.append(a)
```
