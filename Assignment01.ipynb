# VI) Data Visualization

1) Histogram depicting individual column data distribution

```python
graph=data.select_dtypes(include=('int32','int64',float))
graph.hist(figsize=(20,20),color='olive')
plt.show()
```

2) Heatmap for checking the corelations of the data values

```python
corrmatrix=data.corr()
plt.subplots(figsize=(20,15))
sns.heatmap(corrmatrix,vmin=0,vmax=2,annot=True,linewidth=5,cmap='YlGnBu')
```

Here the least correlation is between PL and SW , and maximum between PL and PW

3) CHECKING THE LEAST CORELATION WITH LINE PLOT

```python
x=data['PL']
y=data['SW']
plt.plot(x,y)
plt.title('PL vs SW')
plt.xlabel('PL')
plt.ylabel('SW')
```
